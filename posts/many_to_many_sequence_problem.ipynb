## 0.1. Creating the Dataset

```python
X = list()
Y = list()

X1 = [x1 for x1 in range( 5, 301, 5)]
X2 = [x2 for x2 in range( 20, 316, 5)]
Y = [y for y in range( 35, 331, 5)]

X = np.column_stack( ( X1, X2 ) )
```

```python
X = np.array( X ).reshape( 20, 3, 2 ) # samples, time-steps, features 
Y = np.array( Y ).reshape( 20, 3, 1 ) # samples, time-steps, features
```

# 1.0. Modelling
---

## 1.1. Stacked LSTM

```python
model = Sequential()

model.add( LSTM( 100, activation='relu', input_shape=(3, 2) ) )    # Encoder
model.add( RepeatVector( 3 ) )
model.add( LSTM( 100, activation='relu', return_sequences=True ) ) # Decoder
model.add( TimeDistributed( Dense( 1 ) ) )
model.compile( optimizer='adam', loss='mse' )

# 
history = model.fit( X, Y, epochs=1000, validation_split=0.2, verbose=1, batch_size=3 )
```

```python
X1 = [300, 305, 310]
X2 = [315, 320, 325]

test_input = np.column_stack( ( X1, X2 ) )

test_input = test_input.reshape( ( 1, 3, 2 ) )
print( test_input )
```

```python
test_output = model.predict( test_input, verbose=0 )
print( test_output )
```

## 1.2. Bidirectional LSTM

```python
model = Sequential()

model.add( Bidirectional( LSTM( 100, activation='relu', input_shape=(3, 2) ) ) )
model.add( RepeatVector( 3 ) )
model.add( Bidirectional( LSTM( 100, activation='relu', return_sequences=True ) ) )
model.add( TimeDistributed( Dense( 1 ) ) )

model.compile( optimizer='adam', loss='mse' )

history = model.fit( X, Y, epochs=1000, validation_split=0.2, verbose=1, batch_size=3 )
```

```python
test_output = model.predict( test_input, verbose=0 )
print( test_output )
```
